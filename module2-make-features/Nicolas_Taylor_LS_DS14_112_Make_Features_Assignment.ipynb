```python
#Reads in the data and removes the first row as well as the last two rows

import pandas as pd

df=pd.read_csv('LoanStats_2018Q4.csv', skiprows=1, skipfooter=2)

print(df.shape)
df.head()
```

```python
#Removes the text " months" from every entry in the column "term"

def remove_months(months):
  return (months.str.strip(' months'))

df['term'] = remove_months(df['term'])
df.head()
```

```python
#Changes the data type of the entire "term" column to int

df['term'] = df['term'].astype(int)
df.dtypes[0:10]
```

```python
#Creates a new column that will display a 1 if the loan status is
#Fully Paid or Current and will display a 0 if it isn't

def loan_is_greater(status):
  if status == 'Fully Paid' or status == 'Current':
    return 1
  else:
    return 0

df['loan_status_is_great'] = df['loan_status'].apply(loan_is_greater)

df.head()
```

```python
#Searches for date columns

[col for col in df if col.endswith('_d')]
```

```python


df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)

df.dtypes[40: 50]
```

```python
df['last_pymnt_d'].dt.month
```

```python
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
```

```python
df['last_pymnt_d'].dt.year
```

```python
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
```
